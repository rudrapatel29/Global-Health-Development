## **Analysis & Visualizations**

### How does Current Health Expenditure (CHE) as a percentage of GDP vary across different countries and regions?

```python
che_gdp_by_country = ghed.groupby(['country', 'region'])['che_gdp'].mean().sort_values(ascending=False)
che_gdp_by_country
```

- **Top Health Spenders**: Niue, the United States of America, and Tuvalu are the leading countries in terms of Current Health Expenditure (CHE) as a percentage of GDP, indicating significant investment in health relative to their economies.

- **Regional Distribution**: The Western Pacific Region (WPR) has a high concentration of countries with significant health expenditures, including Niue, Tuvalu, and the Marshall Islands, showcasing regional prioritization of healthcare.

- **Low Health Spendors**: Equatorial Guinea, Bangladesh, and Congo have the lowest CHE as a percentage of GDP, suggesting limited allocation of resources towards healthcare in these regions.

### Domestic General Government Health Expenditure (GGHE-D) as a percentage of GDP by region

```python
gghed_gdp_by_region = ghed.groupby('region')['gghed_gdp'].mean()
gghed_gdp_by_region
```

- **Top Health Spenders**: Europe (EUR) and the Western Pacific Region (WPR) are the leading regions in terms of Domestic General Government Health Expenditure (GGHE-D) as a percentage of GDP, with Europe showing the highest commitment.

- **Regional Distribution**: The Western Pacific Region (WPR) and the Americas (AMR) have significant health expenditure, reflecting strong government involvement in healthcare.

- **Low Health Spenders**: Africa (AFR) and Southeast Asia (SEAR) have the lowest GGHE-D as a percentage of GDP, indicating lower government investment in health relative to their economies.

### How does the average out-of-pocket expenditure (oops_che) on health care vary across different income groups, and what implications might this have for health care affordability in these groups?

### What is the average composition of health expenditure by income group in terms of domestic health expenditure, domestic general government health expenditure, domestic private health expenditure, and external health expenditure?

```python
expenditure_composition = ghed.groupby('income')[['dom_che', 'gghed_che', 'pvtd_che', 'ext_che']].mean()
print(expenditure_composition)
```

- **High Income**: High spending on domestic consumption and general government health expenditure. Minimal spending on external health expenditure.

- **Low Income**: Significant spending on private health services and external health expenditure, with relatively low spending on general government health expenditure.

- **Lower-middle Income**: Balanced spending with notable expenditure on domestic consumption and private health services. Moderate external health expenditure.

- **Upper-middle Income**: High expenditure on domestic consumption and private health services, with relatively lower spending on external health services.

### What is the trend in per capita health expenditure (CHE per capita in USD) over the years in various income groups?

```python
ghed['che_pc_usd'] = pd.to_numeric(ghed['che_pc_usd'], errors='coerce')

che_pc_usd_trend = ghed.groupby(['income', 'year'])['che_pc_usd'].mean().unstack()
che_pc_usd_trend.transpose()

plt.figure(figsize =(13,6))
sns.barplot(che_pc_usd_trend)
```

##### The following observations can be made of the per capita health expenditure (CHE per capita in USD) over the years -

- **High-Income Dominance**: High-income countries consistently lead in health expenditure per capita, reaching $814.50 in 2021.

- **Steady Growth in Upper-Middle-Income**: Upper-middle-income countries show significant growth, with spending increasing from $164.10 in 2000 to $493.37 in 2021.

- **Incremental Rise in Lower-Middle-Income**: Lower-middle-income countries see gradual increases, from $47.84 in 2000 to $178.00 in 2022.

- **Low-Income Struggles**: Low-income countries remain at the bottom, with per capita spending rising only from $15.54 in 2000 to $44.96 in 2021.

- **Pandemic Impact**: Post-2020 data reflects increased health spending across all income groups, driven by the COVID-19 pandemic.

- **Widening Disparity**: The gap between high-income and low-income countries' health expenditure has widened over the years.

- **Resilience in High-Income Countries**: High-income nations maintained or increased spending during economic crises, showcasing resilience.

- **Variability Within Income Groups**: Significant variability in spending exists within the same income groups, influenced by factors like policy and aid.

- **Global Growth Trend**: Health expenditure has consistently increased across all income groups from 2000 to 2021.

- **Future Projections**: If trends continue, upper-middle-income countries may close the spending gap with high-income nations. Significant international support is needed for low-income countries to improve healthcare systems.

### How does the allocation of health expenditure (e.g., Government vs. External) vary between regions, and what factors contribute to these regional disparities?

```python
regional_allocation = ghed.groupby('region')[['dom_che', 'gghed_che', 'pvtd_che', 'ext_che']].mean()
print(regional_allocation)

regional_allocation.plot(kind='bar', figsize=(10, 6), position=0, width=0.8)
plt.title('Regional Allocation of Health Expenditure')
plt.xlabel('Region')
plt.ylabel('Expenditure (% of Current Health Expenditure)')
plt.xticks(rotation=45)
plt.legend(title='Expenditure Type')
plt.tight_layout()
plt.show()
```

Allocation of health expenditure varies significantly across Regions:

- **Africa (AFR):** Relies more on domestic funding (78.76%) but also has a notable reliance on external sources (21.24%).
Americas (AMR): Dominated by domestic funding (97.02%), with minimal external contributions (3.11%).

- **Eastern Mediterranean (EMR):** Similar to AMR, high domestic funding (96.75%) with low external support (3.67%).
Europe (EUR): Highest domestic expenditure (99.19%) and lowest external contributions (1.25%).

- **South-East Asia (SEAR):** Relatively balanced, with a significant portion from external sources (8.04%).

- **Western Pacific (WPR):** Moderate domestic funding (85.28%) and notable external support (15.34%).



Factors contributing to these disparities:

- **Economic Strength:** Wealthier regions like EUR and AMR depend less on external funding.

- **Healthcare Systems:** Stronger government healthcare systems correlate with higher government expenditure (gghed_che).

- **External Support:** Regions with higher external contributions, such as AFR and WPR, might have weaker domestic resources or larger international aid inflows.
